```python
## 필요한 모듈을 임포트한다.
import requests # html
import pdb # 디버깅
import time # 강제로 코드 휴식
from fake_useragent import UserAgent # 가짜 유저 에이전트를 포함하고 있는 모듈
from bs4 import BeautifulSoup # request로 반환 받은 html을 soup 객체로 변환
from pprint import pprint # 결과를 예쁘게 보여주는 모듈
import re
import urllib
import random
```

```python
querys = ["흡연 수명", "광고 매출", "운동 건강", "더위 전기료", "성실 성공"]

all_result = {}

for query in querys:
    
    enc_query = urllib.parse.quote(query)
    
    query_titles = []
    
    pre_title = ""

    i = 0

    while True:

        ua = UserAgent() # 유저 에이전트 객체를 생성한다.

        headers = {'User-Agent': str(ua.random)} # 유저 에이전트를 랜덤하게 생성하고 header 딕셔너리에 담는다.

        # URL 인코딩이 필요한 이유: http://regularmotion.kr/url-encoding-url/

        url = "https://search.naver.com/search.naver?&where=news&query={}&sm=tab_pge&sort=0&photo=0&field=1&reporter_article=&pd=0&ds=&de=&docid=&nso=so:r,p:all,a:t&mynews=0&cluster_rank=147&start={}&refresh_start=0".format(enc_query, 10*i + 1)

        response = requests.get(url, headers=headers)
        
        if response.status_code == 200:
            print("*** 페이지가 정상적으로 로드 되었습니다... ***")

            parsed = response.text

            soup = BeautifulSoup(parsed, "html.parser")

        else:
            print("*** 페이지가 정상적으로 로드 되지 못 하였습니다... ***\n")

            next  

        elements = soup.select("dl > dt > a")
        
        temp_titles = list(map(lambda ele: ele.text, elements))

        cur_title = temp_titles[0]
        
        if pre_title == cur_title: 
            print("그러나 중복된 정보가 수집 되었으므로 다음 query로 넘어가겠습니다.")
            break
        
        query_titles.extend(temp_titles)
        
        print("[{}]   {}번째 페이지까지 완료 되었습니다.    예시:{}\n".format(urllib.parse.unquote(enc_query), i+1, temp_titles[0]))

        pre_title = cur_title

        i += 1
        
        time.sleep(random.randint(5,10))
                  
    all_result[urllib.parse.unquote(enc_query)] = query_titles
```
